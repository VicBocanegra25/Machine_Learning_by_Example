# The logistic Function
* Commonly referred as the **sigmoid function**.
* I maps an input to an output of a value between 0 and 1.
* $\frac{1}{1+\exp({-z})}$

```python
# Visualizing the sigmoid function
import numpy as np
def sigmoid(input):
    return 1.0/(1 + np.exp(-input))
```

```python
# Plotting the function from -8 to 8
import matplotlib.pyplot as plt

# Defining the range
z = np.linspace(-8, 8, 1000)
# Mapping the y values by calling the sigmoid function
y = sigmoid(z)

# Plotting
plt.plot(z, y)
# Drawing lines at critical points (top, middle, bottom)
plt.axhline(y = 0, ls = "dotted", color = "k")
plt.axhline(y = 0.5, ls = "dotted", color = "k")
plt.axhline(y = 1, ls = "dotted", color = "k")
plt.yticks([0.0, 0.25, 0.5, 0.75, 1.0])

plt.xlabel("z")
plt.ylabel("y(z)")

plt.show()
```

The interpretation of this plot is that all input values that are negative
will be classified as 0 and all positive values will be classified as 1.


```python
# Plotting the cost function of logistic regression when y = 1
y_hat = np.linspace (0, 1, 1000)
cost = -np.log(y_hat)
plt.plot(y_hat, cost)
plt.xlabel("Prediction")
plt.ylabel("Cost")
plt.xlim(0, 1)
plt.ylim(0, 7)
plt.show()
```

The cost will be infinite if the model predicts 0 and the correct value was 1.

```python
# Plotting the cost function of logistic regression when y = 1
y_hat = np.linspace (0, 1, 10000)
cost = -np.log(1 - y_hat)
plt.plot(y_hat, cost)
plt.xlabel("Prediction")
plt.ylabel("Cost")
plt.xlim(0, 1)
plt.ylim(0, 7)
plt.show()
```

Inversely, the cost will be infinite if the model predicts 1 when the value was 0.
