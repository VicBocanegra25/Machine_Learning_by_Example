# Implementing K-means Clustering with scikit-learn
* We'll use the Iris dataset

# Step 1: Importing the data


```python
from sklearn import datasets
iris = datasets.load_iris()
# All rows, only the last two columns(features)
X = iris.data[:, 2:4]
y = iris.target
```

# Step 2: Preparing the model

```python
from sklearn.cluster import KMeans
# As before, we'll have 3 clusters (3 types of iris)
kmeans_sk = KMeans(n_clusters = 3, random_state = 42, max_iter = 300, 
                   tol = 1e-4, n_init='auto')
```

```python
# Fit the data: 
kmeans_sk.fit(X,)
```

```python
# Obtaining the clustering results, including clusters for data samples and centroids of individual clusters:
clusters_sk = kmeans_sk.labels_
centroids_sk = kmeans_sk.cluster_centers_
```

# Step 3: Visualizing the data

```python
import matplotlib.pyplot as plt
# Plotting our clusters along with the centroids
plt.scatter(X[:, 0], X[:, 1], c = clusters_sk)
plt.scatter(centroids_sk[:, 0], centroids_sk[:, 1], marker="*",
           s = 200, c="#050505")
plt.show()
```
